```python
import os
import glob
import time
import json
import ephem
import subprocess
from dateutil import parser
from datetime import datetime, timedelta
```

```python
filename = '/Users/ajmendez/Downloads/debug/todays_video.mp4'
```

```python
%%bash -s "$filename" --err ffmpeg
echo "Info for: $1"
ffprobe $1
```

```python
for line in ffprobe.splitlines():
    if any([x in line for x in ['Stream', 'MetaData', 'creation_time', 'handler_name']]):
        print line.replace('(und)',' ').replace('(default)','')
```

```python
item = None
out = []
for line in ffprobe.splitlines():
    if 'Stream' in line:
        item = dict(name=line.replace('(und)',' ').split()[1].replace('#',''))
    elif ('creation_time' in line) and (item is not None):
        item['datestr'] = line.split(' : ')[1].strip()
        item['date'] = parser.parse(item['datestr']) 
        item['ndate'] = int(time.mktime(item['date'].timetuple()))
        out.append(item)
```

```python
TITLE = 'IR {date} Timelapse'
DESCRIPTION='''IR time-lapse video from a Raspberry PI

  Hostname: neon
  Run Time: {runtime}*
  Sunrise: {sunrise}
  Sunset: {sunset}
  delta: {delta:0.2f} seconds
  Captured Time: {captime}
  Note: extracted from archive. Date now matches title.
'''
LAT, LON = '39.331628', '-76.623434' # JHU
TAGS = 'timerasp, timelapse, maryland, baltimore, JHU, "raspberry pi"'

delta = timedelta(minutes=60)
offset = timedelta(hours=5)
here = ephem.Observer()
here.lon, here.lat = LON, LAT

for item in out:
    outfile = os.path.expanduser('~/{ndate}.json'.format(**item))
    if os.path.exists(outfile):
        continue
    d = item['date'] - offset - timedelta(days=1)
    here.date = d-2*delta # ensure that we are before the sunset
    sunrise = ephem.localtime(here.previous_rising(ephem.Sun()))-delta
    sunset = ephem.localtime(here.next_setting(ephem.Sun()))+delta
    delta_time = (sunset-sunrise).total_seconds() / 7200.0
    runtime = int(time.mktime((d-timedelta(days=1)).timetuple()))
    tmp = dict(title=TITLE.format(date=d.strftime('%Y-%m-%d')),
               description=DESCRIPTION.format(sunrise=sunrise,
                                              sunset=sunset,
                                              delta=delta_time,
                                              runtime=runtime,
                                              captime=d),
               tags=TAGS)
    with open(outfile, 'w') as f:
        json.dump(tmp, f, indent=2)
    !SetFile -md '{item["date"].strftime("%m/%d/%Y %H:%M:%S")}' {outfile}
#     break
```
